##if i am solving linear regression problem then i need to divide the independent and dependent features


 Now dividing the abve features into dependent and independent data set

```python
df=pd.read_csv("boston.csv")
df
```

```python
x=df.iloc[:,:-1]
x.head()
```

```python
y=df.iloc[:,-1]
y.head()
```

```python
#linear regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
lin_reg=LinearRegression()
mse=cross_val_score(lin_reg,x,y,scoring='neg_mean_squared_error',cv=5)# no mean square present
print(mse) 
mean_mse=np.mean(mse)
print(mean_mse)
```

```python
#ridge regression
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
ridge=Ridge()
#params={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20]}
#add more value
params={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
#ridge_regressor=GridSearchCV(ridge, params,scoring='neg_mean_squared_error',cv=5)
#change cv to 10
ridge_regressor=GridSearchCV(ridge, params,scoring='neg_mean_squared_error',cv=10)
ridge_regressor.fit(x,y)#fit is used for relation ship
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

```python
#Lasso regression
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso=Lasso()
params={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20]}
lasso_regressor=GridSearchCV(lasso, params,scoring='neg_mean_squared_error',cv=5)
lasso_regressor.fit(x,y)
```

```python
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

```python
#train and test split (# to estimate the performance of the ml algorithm) or to improve the score
from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test = train_test_split(
x, y, test_size=0.33, random_state=42) #import from sklearn page example and test size .33 means test will be 33% and train data will hae 70 % and random state value can be anything
```

```python
#linear regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
lin_reg=LinearRegression()
lin_reg.fit(x_train,y_train)
mse=cross_val_score(lin_reg,x_train,y_train,scoring='neg_mean_squared_error',cv=5)# no mean square present
print(mse) 
mean_mse=np.mean(mse)
print(mean_mse)
#value should go towards 0 i.e. performance is better
```

```python
#ridge regression
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
ridge=Ridge()
#params={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20]}
#add more value
params={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
#ridge_regressor=GridSearchCV(ridge, params,scoring='neg_mean_squared_error',cv=5)
#change cv to 10
ridge_regressor=GridSearchCV(ridge, params,scoring='neg_mean_squared_error',cv=10)
ridge_regressor.fit(x_train,y_train)
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

#improvement is quite bad as we are not going towards zero
